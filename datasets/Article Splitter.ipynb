```python
import pandas as pd
```

```python
df = pd.read_csv("articles.csv")
df
```

```python
categories = ["poder","mercado","esporte","tec","educacao","turismo"]
limit = 2000
with open("articles-filtered-{}.tsv".format(limit),'w') as f:
    f.write("text\tcategory\n")
    for category in categories:
        news = df.loc[df["category"] == category][["text","category"]][0:limit]
        for row in news.iterrows():
            f.write('{}\t{}\n'.format(row[1].get("text").replace("\t"," "), row[1].get("category")))
        
```
